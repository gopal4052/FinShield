```python
# ------------------------------------------------------------
# Validate update events
# Expected: matched_update_accounts = update_events
# ------------------------------------------------------------

spark.sql("""
SELECT COUNT(*) AS update_events
FROM data_engineering.table_metadata.account_cdc_events_audit
WHERE operation = 'U'
""").display()

spark.sql("""
SELECT COUNT(*) AS matched_update_accounts
FROM data_engineering.table_metadata.account_cdc_events_audit e
JOIN data_engineering.silver_layer.dim_accounts a
ON e.account_id = a.account_id
WHERE e.operation = 'U'
AND a.status = e.status
AND a.credit_score = e.credit_score
AND a.kyc_verified = e.kyc_verified
AND a.is_deleted = false
""").display()
```

```python
spark.sql("""
SELECT
    status,
    is_deleted,
    COUNT(*) AS account_count
FROM data_engineering.silver_layer.dim_accounts
WHERE is_deleted = true
AND status <> 'CLOSED'
GROUP BY status, is_deleted
""").display()
```

```python
spark.sql("""
SELECT COUNT(*) AS closed_deleted_accounts
FROM data_engineering.silver_layer.dim_accounts
WHERE is_deleted = true
AND status = 'CLOSED'
""").display()
```

```python
spark.sql("""
SELECT
    status,
    is_deleted,
    COUNT(*) AS account_count
FROM data_engineering.silver_layer.dim_accounts
GROUP BY status, is_deleted
ORDER BY status, is_deleted
""").display()
```

```python
spark.sql("""
SELECT
    COUNT(*) AS total_accounts,
    COUNT(DISTINCT account_id) AS distinct_accounts
FROM data_engineering.silver_layer.dim_accounts
""").display()
```

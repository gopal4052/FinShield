Control Table

```python
spark.sql("""
CREATE TABLE IF NOT EXISTS data_engineering.table_metadata.finshield_control (
    pipeline_name STRING,
    source_name STRING,
    last_watermark BIGINT,
    updated_at TIMESTAMP
)
USING DELTA
""")
```

```python
spark.sql("""
INSERT INTO data_engineering.table_metadata.finshield_control
SELECT
    'silver_transactions_incremental' AS pipeline_name,
    'paysim_transactions' AS source_name,
    0 AS last_watermark,
    current_timestamp() AS updated_at
""")
```

```python
spark.sql("""
SELECT *
FROM data_engineering.table_metadata.finshield_control
""").display()
```

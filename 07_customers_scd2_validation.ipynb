```python
print("Check one current record per account:")

spark.sql(f"""
SELECT
    account_id,
    COUNT(*) AS current_record_count
FROM {DIM_CUSTOMERS_TABLE}
WHERE is_current = true
GROUP BY account_id
HAVING COUNT(*) > 1
""").display()


print("Customers with history:")

spark.sql(f"""
SELECT
    account_id,
    COUNT(*) AS version_count
FROM {DIM_CUSTOMERS_TABLE}
GROUP BY account_id
HAVING COUNT(*) > 1
LIMIT 20
""").display()


print("Sample SCD2 history:")

spark.sql(f"""
SELECT *
FROM {DIM_CUSTOMERS_TABLE}
WHERE account_id IN (
    SELECT account_id
    FROM {DIM_CUSTOMERS_TABLE}
    GROUP BY account_id
    HAVING COUNT(*) > 1
    LIMIT 5
)
ORDER BY account_id, effective_start_date
""").display()
```
